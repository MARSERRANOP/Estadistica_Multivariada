```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA

# Datos
X = [[0, 0], [0, 1], [2, 1], [3, 4], [5, 2]]
X_df = pd.DataFrame(data=X, columns=['Coordenada x', 'Coordenada y'])

# Etiquetas
labels = ['P1', 'P2', 'P3', 'P4', 'P5']
X_df['Label'] = labels

# Dividir los datos en caractersticas y etiquetas
X = X_df.drop(columns=['Label'])
y = X_df['Label']

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Escalar las caractersticas
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Imprimir las formas de los conjuntos de datos
print("X_train.shape, y_train.shape:", X_train.shape, y_train.shape)
print("X_test.shape, y_test.shape:", X_test.shape, y_test.shape)

# Visualizacin original
plt.scatter(X_df.iloc[:, 0], X_df.iloc[:, 1], marker="o", alpha=0.8, color='red')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Coordenada x', fontsize=15)
plt.ylabel('Coordenada y', fontsize=15)
plt.grid()
plt.show()

# Aplicar PCA
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print("PCA train shape:", X_train_pca.shape)
print("PCA test shape:", X_test_pca.shape)

# Visualizacin PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_train[:, 0], X_train[:, 1], alpha=0.8, label='Original', color='red')
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], alpha=0.8, label='PCA', color='blue')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Coordenada x', fontsize=15)
plt.ylabel('Coordenada y', fontsize=15)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
